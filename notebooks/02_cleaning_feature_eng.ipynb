# 02 - Cleaning and Feature Engineering
**Project:** Social Media Advertising Analytics  
**Goal:** Handle edge cases and engineer three derived KPIs.

---

```python
import pandas as pd
import numpy as np

df = pd.read_csv('../data/social_media_ads.csv', parse_dates=['Date'])
print(f'Loaded: {df.shape[0]:,} rows, {df.shape[1]} columns')
```

## 1. Check for zero-value edge cases

```python
zero_clicks      = (df['Clicks'] == 0).sum()
zero_impressions = (df['Impressions'] == 0).sum()
zero_duration    = (df['Duration'] == 0).sum()

print(f'Zero Clicks:      {zero_clicks}')
print(f'Zero Impressions: {zero_impressions}')
print(f'Zero Duration:    {zero_duration}')

df['Clicks']      = df['Clicks'].replace(0, np.nan)
df['Impressions'] = df['Impressions'].replace(0, np.nan)
df['Duration']    = df['Duration'].replace(0, np.nan)

print('\nZero values replaced with NaN for safe division.')
```

## 2. Engineer derived KPIs

```python
df['CTR']            = df['Clicks'] / df['Impressions']
df['Cost_per_Click'] = df['Acquisition_Cost'] / df['Clicks']
df['ROI_per_Day']    = df['ROI'] / df['Duration']

print('New features:')
print(df[['CTR', 'Cost_per_Click', 'ROI_per_Day']].describe().round(4))
```

## 3. Extract temporal features

```python
df['day_name']   = df['Date'].dt.day_name()
df['month_name'] = df['Date'].dt.month_name()
df['quarter']    = df['Date'].dt.quarter

print(df[['Date', 'day_name', 'month_name', 'quarter']].head())
```

## 4. Save cleaned dataset

```python
df.to_csv('../data/social_media_ads_clean.csv', index=False)
print(f'Saved: {df.shape[0]:,} rows, {df.shape[1]} columns')
```
